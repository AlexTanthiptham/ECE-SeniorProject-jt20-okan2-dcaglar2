```python
idealStepMotorSpeed =30 #rpm
gearRatio = 2 #1:2
sensorTurnRate = idealStepMotorSpeed*gearRatio
sensorRefreshRate = 100 #Hz
sensorTurnPeriod = 1/sensorTurnRate *60
sensorMeasureRadiusData =[2.5,2.5]#[3.5,7,10.5,14,17.5,21,24.5,28] #1-8m in terms of cm
rayCount = int(sensorTurnPeriod*sensorRefreshRate)
print(f"it takes {sensorTurnPeriod}s for 1 rotation ")
print(f"In 1 rotation we get {sensorTurnPeriod*sensorRefreshRate}")
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as pch
import math
n=210

radii = [100,200]
colours =["red","green","blue"]
figure, axes = plt.subplots()
def drawSensorRayCast(r,colour):
    # Drawing_colored_circle = plt.Circle(( 0,0 ), r )
    # axes.add_artist( Drawing_colored_circle )
    c = 2*math.pi/rayCount
    # lastX,lastY = 0,0
    x,y =0,0
    radius = sensorMeasureRadiusData[int(r/100)-1]
    for i in range(rayCount):
        lastX = x
        lastY = y
        x = r * math.cos(i*c) 
        y = r * math.sin(i*c)
        circle= plt.Circle(( x,y ), sensorMeasureRadiusData[int(r/100)-1] ,color = colour)
        line = pch.FancyArrow(0,0,x,y ,color =colour)
        axes.add_artist( circle )
        axes.add_artist( line)
    print(f"For {int(r/100)}: distance is: {math.sqrt((x-lastX-radius)**2+(y-lastY-radius)**2)} radius:{radius}")

for c,d in enumerate(radii):
    drawSensorRayCast(d,colours[c])
plt.xlim([-n,n])
plt.ylim([-n,n])

# plt.title( 'Colored Circle' )
plt.show()
```
